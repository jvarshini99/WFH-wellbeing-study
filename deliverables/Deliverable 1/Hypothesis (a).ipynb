Hypothesis: Participantsâ€™ age will negatively correlate with financial and material stability (the last two questions on the Flourishing Scale) 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the dataset
demographic_df = pd.read_csv("Demographic.csv")
Friday_PM_3_Month_df = pd.read_csv("Friday_PM_3_Month.csv")

# Rename columns in df and merge
demographic_df = demographic_df.rename(columns={"Mobile ID": "mbl_cod", "Age ": "Age"})
merged_df = pd.merge(demographic_df, Friday_PM_3_Month_df, on="mbl_cod")

# Select the relevant columns and drop 0 and NaNs
financial_stability_df = merged_df[["Age", "LIVING_EXPENSES"]].dropna(subset=["LIVING_EXPENSES"])

# Remove NaN and infinite values from the age and financial stability arrays
age = np.array(financial_stability_df["Age"].dropna().replace([np.inf, -np.inf], np.nan).dropna())
financial_stability = np.array(financial_stability_df["LIVING_EXPENSES"].dropna().replace([np.inf, -np.inf], np.nan).dropna())

# Create scatter plots
plt.scatter(age, financial_stability, color="red", label="Financial Stability")
plt.title("Age vs. Financial Stability")
plt.xlabel("Age")
plt.ylabel("Financial and Material Stability")
plt.legend()
plt.show()

# Calculate the correlation coefficients and p-values
corr1, p_value1 = pearsonr(age, financial_stability)

# Print the correlation coefficients and p-values
print(f"Correlation between age and financial stability: {corr1:.2f}, p-value: {p_value1:.2f}")
```
