Demographic Data Graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the dataset
demographic_df = pd.read_csv("Demographic.csv")

# Create bar chart of state
state_counts = demographic_df["State"].value_counts()
plt.bar(state_counts.index, state_counts.values)
plt.title("State Bar Chart")
plt.xlabel("State")
plt.ylabel("Number")
plt.gcf().set_size_inches(12, 6)
plt.show()

# Create bar chart of age
age_counts = demographic_df["Age "].value_counts()
plt.bar(age_counts.index, age_counts.values)
plt.title("Age Bar Chart")
plt.xlabel("Age")
plt.ylabel("Number")
plt.gcf().set_size_inches(12, 6)
plt.show()

# Create box plot of age
plt.boxplot(demographic_df["Age "].dropna())
plt.title("Age Box Plot")
plt.ylabel("Age")
plt.show()

# Create bar chart of industry
industry_counts = demographic_df["Industry"].value_counts()
plt.bar(industry_counts.index, industry_counts.values)
plt.title("Industry Bar Chart")
plt.xlabel("Industry")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number")
plt.gcf().set_size_inches(12, 6)
plt.show()
```
