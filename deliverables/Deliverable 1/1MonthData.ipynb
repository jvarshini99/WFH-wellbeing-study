### **Analysing the data**

**DailyAM Data**

*Number of Breaks taken*

```python
#Finding the average number of breaks taken by each rsp_id
df_avgbreaks = df_month.groupby('rsp_id')['DAILY_BREAKS'].mean().reset_index()
df_avgbreaks.head(10)
```

```python
#plotting average breaks on a graph
df_avgbreaks.plot(x='rsp_id', y='DAILY_BREAKS', kind='bar', figsize=(13, 6))
plt.xlabel('rsp_id')
plt.ylabel('Average DAILY_BREAKS')
plt.title('Average DAILY_BREAKS by rsp_id')
plt.show()
```

**FridayAM Data**

*Florence score*

```python
#make a copy of the df
dfFriFlo = dfFridayAM.copy()
```

```python
#Florence score = LIFE_SATISFACTION + HAPPINESS + PHYSICAL_HEALTH + MENTAL_HEALTH + WORTHWHILE + PURPOSE + DELAYED_HAPPINESS +  PROMOTE_GOOD + CONTENT_RELATIONSHIPS +  SATISFYING_RELATIONSHIPS +  LIVING_EXPENSES + FOOD_HOUSING
dfFriFlo["florence_score"]=dfFriFlo["LIFE_SATISFACTION"]+dfFriFlo["HAPPINESS"]+dfFriFlo["PHYSICAL_HEALTH"]+dfFriFlo["MENTAL_HEALTH"]+dfFriFlo["WORTHWHILE"]+dfFriFlo["PURPOSE"]+dfFriFlo["PROMOTE_GOOD"]+dfFriFlo["DELAYED_HAPPINESS"]+dfFriFlo["CONTENT_RELATIONSHIPS"]+dfFriFlo["SATISFYING_RELATIONSHIPS"]+dfFriFlo["LIVING_EXPENSES"]+dfFriFlo["FOOD_HOUSING"]
```

```python
dfFriFlo.head(10)
```

```python
#Average florence score calculated for each rsp_id
df_avgflorence = dfFriFlo.groupby('rsp_id')['florence_score'].mean().reset_index()
df_avgflorence.head(10)
```

```python
#plotting the average florence score
df_avgflorence.plot(x='rsp_id', y='florence_score', kind='bar', figsize=(13, 6))
plt.xlabel('rsp_id')
plt.ylabel('Average florence score')
plt.title('Average florence score by rsp_id')
plt.show()
```

*Merging the two dataframes wrt rsp_id*

```python
#Merge df_avgflorence and df_avgbreaks
df_merged=pd.merge(df_avgflorence, df_avgbreaks, on='rsp_id', how="right")
print(df_merged)
```

```python
#creating a plot for the merged dataframe df_merged on Daily breaks and florence score
plt.scatter(df_merged['DAILY_BREAKS'], df_merged['florence_score'])
plt.xlabel("Average Daily Breaks") #x label
plt.ylabel("Florence Score") #y label
plt.title("1 Month Data")
plt.show()
```

## We can see from the graph that people who took an average of daily breaks in the range of 2 to 5 have a high florence score. This is based of the 1 month data extracted.
