Hypothesis: Based on question #15 in the  Computer Workstation Checklist (with 4 responses regarding ergonomics training), participants with lower scores will report less pain at 6-months. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the dataset
workstation_df = pd.read_csv("Computer_Workstations_6_Month.csv")
Friday_AM_6_Month_df = pd.read_csv("Friday_AM_6_Month.csv")

# Merge DF
df = pd.merge(workstation_df, Friday_AM_6_Month_df, on="mbl_cod")

# Select relevant columns and drop any NaN values
relevant_columns = ["POSTURE_TRAINING", "METHODS_TRAINING", "WMSD_SIGNS", "WORKSTATION_ADJUSTMENT", "PHYSICAL_HEALTH"]
df = df[relevant_columns].dropna()

# Loop through the four columns and create a bar chart for each one
for column in ["POSTURE_TRAINING", "METHODS_TRAINING", "WMSD_SIGNS", "WORKSTATION_ADJUSTMENT"]:
    column_groups = df.groupby(column)["PHYSICAL_HEALTH"].mean()
    plt.bar(column_groups.index, column_groups.values, width=0.4)
    plt.xticks(column_groups.index)
    plt.xlabel(column)
    plt.ylabel("Average PHYSICAL_HEALTH Score")
    plt.title(f"Average PHYSICAL_HEALTH Score by {column}")
    plt.show()

# Calculate the correlation between the four columns and PHYSICAL_HEALTH
correlations = df[["POSTURE_TRAINING", "METHODS_TRAINING", "WMSD_SIGNS", "WORKSTATION_ADJUSTMENT", "PHYSICAL_HEALTH"]].corr(method="pearson")
# Print the correlation coefficients
print("Correlations: ")
print(correlations["PHYSICAL_HEALTH"])
```
