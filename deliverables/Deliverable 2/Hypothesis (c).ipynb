Hypothesis: Participants working in healthcare will have lower mental health scores on the Flourishing Scale than those working in other industries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the dataset
demographic_df = pd.read_csv("Demographic.csv")
Friday_PM_3_Month_df = pd.read_csv("Friday_PM_3_Month.csv")

# Rename columns in df and merge
demographic_df = demographic_df.rename(columns={"Mobile ID": "mbl_cod", "Age ": "Age"})
merged_df = pd.merge(demographic_df, Friday_PM_3_Month_df, on="mbl_cod")

# Select the relevant columns and drop 0 and NaNs
mental_health_df = merged_df[["Industry", "MENTAL_HEALTH"]].dropna(subset=["MENTAL_HEALTH"])

# Group data by industry and calculate the average mental health score
avg_mental_health_by_industry = mental_health_df.groupby("Industry")["MENTAL_HEALTH"].mean()

# Plot the average mental health score by industry
plt.bar(avg_mental_health_by_industry.index, avg_mental_health_by_industry.values)
plt.title("Average Mental Health Score by Industry")
plt.xlabel("Industry")
plt.xticks(rotation=45, ha='right')
plt.gcf().set_size_inches(12, 6)
plt.ylabel("Average Mental Health Score")
plt.show()
```
