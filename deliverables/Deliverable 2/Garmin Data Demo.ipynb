```python
import pandas as pd
import matplotlib.pyplot as plt

demo_file_0 = 'garmin.17180706.csv'
demo_file_1 = 'garmin.11822993 2.csv'

# Replace 'filename.csv' with the actual filename and path of your CSV file.
df = pd.read_csv(demo_file_0)

# Drop rows with 0 value in the 'val' column
df = df[df['val'] != 0]

# Analysis 1: group value according to data type
# Group the data by 'data_type' and calculate the mean, median, and standard deviation of 'val' for each group
grouped = df.groupby('data_type')['val'].agg(['mean', 'median', 'std'])

# Create a bar chart of the mean values for each data type
grouped['mean'].plot(kind='bar')

# Create a boxplot of the 'val' column for each data type
df.boxplot(column='val', by='data_type')

# Rotate the x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Show the plots
plt.show()


# Analysis 2: inspect data type changes over date and time
# Convert the 'dte_tme' column to a Pandas datetime format
df['dte_tme'] = pd.to_datetime(df['dte_tme'])

# Create a larger plot with 12 inches in width and 6 inches in height
plt.figure(figsize=(12, 6))

# Set the 'dte_tme' column as the index of the DataFrame
df.set_index('dte_tme', inplace=True)

# Create a line plot of 'val' against 'dte_tme' for each 'data_type'
for data_type, group in df.groupby('data_type'):
    group['val'].plot(label=data_type)

# Add a title and axis labels
plt.title('Time Series Plot')
plt.xlabel('Date and Time')
plt.ylabel('Value')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```
